```python
#Divisão de Telas(Subgráficos)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('/content/trees.csv')
base.head()
```

```python
#Girth com Volume
plt.scatter(base.Girth, base.Volume)
```

```python
#Girth com heigth
plt.scatter(base.Girth, base.Height)
```

```python
#Height com Volume
plt.scatter(base.Height, base.Volume, marker='*')
```

```python
plt.hist(base.Volume)
```

```python
#Imprimindo os gráficos juntos
plt.figure(2)

plt.suptitle('Árvores', fontsize=16)

plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.ylabel('Circunferência')
plt.xlabel('Volume')
plt.title('Dispersão da Circunferência vs Volume', fontsize=10)

plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.ylabel('Circunferência')
plt.xlabel('Altura')
plt.title('Dispersão da Circunferência vs Altura', fontsize=10)

plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume, marker='*')
plt.xlabel('Altura')
plt.ylabel('Volume')
plt.title('Dispersão da Altura vs Volume', fontsize=10)

plt.subplot(2,2,4)
plt.hist(base.Volume)
plt.xlabel('Frequência')
plt.ylabel('Volume')
plt.title('Histograma do Volume', fontsize=10)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```
