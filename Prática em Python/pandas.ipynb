```python
#renomear colunas
dados.rename(columns={'duration':'duração', 'purpose':'propósito'})
```

```python
#renomeação não é persistente
dados.head(5)
```

```python
#para a renomeação ser persistida, deve-se passar o parâmetro inplace=True
dados.rename(columns={'duration':'duração', 'purpose':'propósito'}, inplace=True)
```

```python
#excluir coluna
dados.drop('checking_status', axis=1, inplace=True)
print(dados)
```

```python
#verificar dados nulos
dados.isnull()
```

```python
#verificar dados nulos
dados.isnull().sum()
```

```python
#retirar todas colunas com NaN
dados.dropna()
```

```python
#preencher dados faltantes
dados['duração'].fillna(0, inplace=True)
```
